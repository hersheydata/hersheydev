## Lambda Function (익명 함수)

파이썬에서는 람다 함수를 통해 이름이 없는 익명의 함수를 만들 수 있다.

람다 함수의 가장 큰 장점은 코드의 간결함, 메모리의 절약이라고 볼 수 있다.

람다 함수는 lambda라는 키워드로 생성할 수 있다.

위 구조처럼, 결과 부분을 return 키워드없이 자동으로 return 해준다.

또한, 익명 함수라는 이름에 걸맞게 함수의 이름을 별도로 지정하지도 않는다.

```python
def summation(x):
    x + 1
```

```python
lambda x: x+1
```

위 람다 함수는 인자로 받은 x에 +1을 해주는 함수로, 위 summation 함수와 완벽히 동일한 기능을 수행한다. 그리고 정의와 동시에 아래와 같이 응용하여 사용이 가능하다

```python
(lambda x:x+1)(3) #x에 3을 받은 경우의 결과값을 리턴함
```

하지만 함수에 이름이 없고, 저장된 변수가 없기 때문에 재사용할 수는 없음.

그러나, 람다 함수 자체도 객체이기 때문에 정의와 동시에 변수에 할당(바인딩)할 수는 있음

```python
func = lambda x : x+1
func(4)
```

### lambda 함수 사용 예시

python 내장 정렬함수인 sorted의 경우, key 위치 인자에 함수를 보내서 함수에서 지정한 결과값에 따라 정렬을 할 수 있다.

```python
target = ['  cat ', ' tiger ', '    dog', 'snake   ']
```

위 알파벳 순서가 아니라, 공백을 제외한 문자의 길이를 기준으로 정렬을 하고 싶다면 정렬의 기준(key)으로 사용할 값을 리턴하는 함수를 생성(strip 메서드 사용)하고 sorted 함수에 넘겨주면 된다.

```python
def my_key(string):
    return len(string.strip()) #.strip() : 공백 제거 메서드

target = ['  cat ', ' tiger ', '    dog', 'snake   ']
print(sorted(target, key=my_key))
```

```python
my_key("guess_key")
```

그러나 my_key라는 함수는 이번 정렬만을 위한 일회성 함수이기 때문에, 재사용할 이유가 없다면 훨씬 빠른 람다 함수를 사용할 수 있다.

```python
target = ['  cat ', ' tiger ', '    dog', 'snake   ']
print(sorted(target, key=lambda x:len(x.strip())))
```
