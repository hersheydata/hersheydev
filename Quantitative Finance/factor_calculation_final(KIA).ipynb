# Factor Calculation

## 1. BM Ratios
### Book to Market
* $Common\,Shareholders'\,Equity\,/\,Market\,Cap$

-------
## 2. EP Ratios
### PER
* $Price\,per\,Share\,/EPS=Market\,Cap\,/Net\,Income$
 
### PBR
* $Price\,per\,Share\,/\,BPS=Market\,Cap\,/\,Total \,Equity$

### PSR
* $Price\,per\,Share\,/\,SPS=Market\,Cap\,/\,Sales$

### EV/EBITDA
* $Enterprise\,Value\,/\,EBITDA = (Equity\,Value + Net\, Debt)/EBITDA$


```python
from marcap import marcap_data

# 특정 기간 단일 종목
df = marcap_data('2008-12-31', '2020-12-31', code='000270')
```

```python
df = df[["Marcap"]]
marcap = df.iloc[::229,:]
index = data.index

marcap = marcap.set_index(index)
code = pd.Series("000270", index = index)
code = pd.DataFrame(code)

code = code.rename(columns={0:"code"})
code = code.set_index(index)
```

```python
data = pd.concat([code, marcap, data], axis=1)
```

```python
# Market Factors
data["BM"] = data["보통주자본금"]/data["Marcap"]
data["PER"] = data["Marcap"]/data["당기순이익"]
data["PBR"] = data["Marcap"]/data["자본총계"]
data["PSR"] = data["Marcap"]/data["매출액"]
data["EV/EBIT"] = (data["Marcap"]+(data["단기차입금"]+data["장기차입금"]-data["현금및현금성자산"]))/data["영업이익"]
```

```python
# Profitability
data["ROE"] = data["당기순이익"]/data["자본총계"]
data["ROA"] = data["당기순이익"]/data["자산총계"]
data["GP"] = data["매출총이익"]/data["매출액"]
data["ROTC"] = data["영업이익"]/data["보통주자본금"]
data["ROIC"] = data["영업이익"]/((data["유동자산"]-data["유동부채"])+data["유형자산"])
data["RNOA"] = data["당기순이익"]/((data["매출채권"]+data["유형자산"]+data["재고자산"]+data["현금및현금성자산"])-(data["부채총계"]-data["단기차입금"]-data["장기차입금"]))
```

### ROE (Return on Equity)
- $Net\,Income\,/\,Total\,Equity$

### ROA (Return on Asset)
- $Net\,Income\,/\,Total\,Asset$

### Gross Profit Ratio
- $Gross\,Profit\,/\,Sales=(Sales-COGS)/Sales$

where COGS stands for Cost of Goods Sold

### ROTC (Return on Total Capital)
- $NOPAT\,/\,Total\,Capital = EBIT(1-t)\,/\,Total\,Capital$

분자의 경우, EBIT를 써도 무방

### ROIC (Return on Invested Capital)
- $NOPAT\,/\,Invested\,Capital=EBIT(1-t)\,/\,Invested\,Capital$

분자의 경우, EBIT를 써도 무방

### RNOA (Return on Net Operating Asset)
- $Net\,Income/NOA\,=\,Net\,Income\,/(Operating\,Asset - Operating\,Liabilities)\,=\,Net\,Income\,/ (Operating\,Assets - (Total\, Liabilities - IBD))$

where IBD stands for "Interest Bearing Debts (LTD + STD)"

-------


### cf) Core components for multiple valuation

* Enterprise Value = Equity Value + Net Debt
* Equity Value = Net Income * PER 
* Net Debt = IBD - C&CE = (LTD + STD) - (Cash + Investments)
